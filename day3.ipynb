```python
import numpy as np
from itertools import dropwhile
import matplotlib.pyplot as plt
```

```python
with open("day3.txt", "r") as f:
    text = f.read().split("\n")[:-1]
```

```python
def filter_func(char_list, n):
    if n == 1:
        return max(char_list)
    dig = max(char_list[:-n+1])
    char_list = list(dropwhile(lambda x: x < dig, char_list))[1:]
    res = dig + filter_func(char_list, n - 1)
    return res

res1 = sum([int(filter_func([c for c in x], 2)) for x in text])
res2 = sum([int(filter_func([c for c in x], 12)) for x in text])
```
