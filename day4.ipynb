```python
import numpy as np
from scipy.signal import convolve2d
```

```python
with open("day4.txt", "r") as f:
    text = f.read().split("\n")[:-1]
text_int = np.array([[1 if c == '@' else 0 for c in x] for x in text])
```

```python
kernel = np.ones((3, 3))
mat = convolve2d(text_int, kernel, mode="same") * text_int
res1 = sum([mat == i for i in [1, 2, 3, 4]]).astype(bool)

res2 = 0
keep_run = True
while keep_run:
    mat = convolve2d(text_int, kernel, mode="same") * text_int
    res1 = sum([mat == i for i in [1, 2, 3, 4]]).astype(bool)
    if res1.sum() == 0:
        keep_run = False
    res2 += res1.sum()
    text_int[res1] = 0
```
