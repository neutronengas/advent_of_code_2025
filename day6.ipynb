```python
import numpy as np
import re
```

```python
with open("day6.txt", "r") as f:
    text = f.read().split("\n")[:-1]
```

```python
numbs = text[:-1]
ops = text[-1]

pattern = "\d+"
arrs = []
for string in numbs:
    match = re.findall(pattern, string)
    arrs.append([int(i) for i in match])
arrs = np.array(arrs).T

ops = [c for c in ops if c in ['*', '+']]
prod_idx = np.array(ops) == '*'
res1 = np.prod(arrs[prod_idx], axis=1).sum() + arrs[~prod_idx].sum()

numbs_T = np.array([[c for c in row] for row in numbs]).T
new_blocks = []
block = []
for row in numbs_T:
    string = "".join(row).replace(" ", "")
    if string == "":
        new_blocks.append(block)
        block = []
    else:
        block.append(int(string))
new_blocks.append(block)

arrs2 = new_blocks
arrs2_prod = [new_blocks[i] for i, x in enumerate(ops) if x == '*']
arrs2_sum = [new_blocks[i] for i, x in enumerate(ops) if x == '+']
res2 = sum([np.prod(arr) for arr in arrs2_prod]) + sum([sum(arr) for arr in arrs2_sum])
```
