```python
import numpy as np
import re
```

```python
with open("day2.txt", "r") as f:
    text = f.read()
```

```python
inp_spl = text.split(",")
inp_spl_int = [[int(i) for i in x.split("-")] for x in inp_spl]
arrs = np.concatenate([np.arange(x[0], x[1] + 1) for x in inp_spl_int], axis=0)
pattern1 = r"^(?:(.)\1|(..)\2|(...)\3|(....)\4|(.....)\5)$"
pattern2 = r"^(?:(.)\1+|(..)\2+|(...)\3+|(....)\4+|(.....)\5)$"
res1 = sum(arrs[np.array([bool(re.fullmatch(pattern1, str(x))) for x in arrs])])
res2 = sum(arrs[np.array([bool(re.fullmatch(pattern2, str(x))) for x in arrs])])
```
