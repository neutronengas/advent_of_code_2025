```python
import numpy as np
```

```python
with open("day1.txt", "r") as f:
    text = f.read().split("\n")[:-1]
```

```python
def map_func(s):
    res = int(s[1:])
    if s[0] == 'R':
        res *= -1
    return res

def transform_text(text):
    new_text = []
    for seq in text:
        letter = seq[0]
        for _ in range(int(seq[1:])):
            new_text.append(letter + "1")
    return new_text

data = np.array([map_func(s) for s in text])
data2 = np.array([map_func(s) for s in transform_text(text)])
res1 = (((data.cumsum() + 50) % 100) == 0).sum()
res2 = (((data2.cumsum() + 50) % 100) == 0).sum()
```
