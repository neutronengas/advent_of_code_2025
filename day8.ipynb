```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("day8.txt", "r") as f:
    text = f.read().split("\n")[:-1]
```

```python
vecs = np.array([[int(i) for i in x.split(",")] for x in text])
l = len(vecs)

dist = np.linalg.norm(vecs[:, None, :] - vecs[None, :, :], axis=-1)
dist[dist == 0] = dist.max()

# for _ in range(1000):
ind = np.array(np.unravel_index(np.argsort(dist.flatten()), (l, l))).T
new_cables = ind[:2*10]
cables_mat = np.zeros((l, l))
cables_mat[new_cables[:, 0], new_cables[:, 1]] = 1 
adj_mat = (cables_mat + np.eye(l)).astype(int)

while True:
    new_adj_mat = adj_mat @ adj_mat
    new_adj_mat[new_adj_mat > 0] = 1
    if (new_adj_mat - adj_mat).sum() == 0:
        break
    adj_mat = new_adj_mat

occs, lengths = np.unique(np.sort(adj_mat.sum(axis=0)), return_counts=True)
res1 = np.prod(np.repeat(occs, lengths // occs)[-3:])
res1
```
