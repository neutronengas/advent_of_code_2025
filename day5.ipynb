```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("day5.txt", "r") as f:
    text = f.read()
```

```python
ranges, ints = text.split("\n\n")
mins, maxs = np.array([[int(x.split("-")[0]), int(x.split("-")[1])] for x in ranges.split("\n")]).T
ints = np.array([int(i) for i in ints.split("\n")[:-1]])
res1 = ((ints[:, None] >= mins[None, :]) * (ints[:, None] <= maxs[None, :])).max(axis=1).sum()

def has_ovlp(int1, int2):
    if min(int1) > min(int2):
        return has_ovlp(int2, int1)
    return max(int1) >= min(int2)

intervals = np.stack([mins, maxs], axis=1)
adj_mat = np.array([[has_ovlp(int1, int2) for int1 in intervals] for int2 in intervals]).astype(float)
while True:
    adj_mat_new = adj_mat @ adj_mat
    adj_mat_new[adj_mat_new > 0] = 1. 
    if (adj_mat_new - adj_mat).sum() == 0:
        break
    adj_mat = adj_mat_new

res2 = []
for row in adj_mat:
    add = (intervals[np.nonzero(row)].max() - intervals[np.nonzero(row)].min() + 1) / row.sum()
    res2.append(add)
res2 = sum(res2)
```
