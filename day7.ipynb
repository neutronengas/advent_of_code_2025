```python
import numpy as np
```

```python
with open("day7.txt", "r") as f:
    text = f.read().split("\n")[:-1]
```

```python
mat = np.array([[1 if c == 'S' or c == '^' else 0 for c in row] for row in text])
beams = mat[0]
active_splits = 0
for row in mat[2:]:
    active_splits += (beams * row).sum()
    new_beams = np.stack([np.concatenate([[0], row[:-1]]), np.concatenate([row[1:], [0]])], axis=1).max(axis=1)
    beams = np.stack([beams * (1 - row), new_beams], axis=1).max(axis=1)
res1 = active_splits

reachable = mat[0]
for row in mat[1:]:
    reachable = reachable * (1 - row) + np.concatenate([[0], (row * reachable)[:-1]]) + np.concatenate([(row * reachable)[1:], [0]])
res2 = reachable.sum()
```
